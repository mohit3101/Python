### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
x1=float(input('Enter the x-coordinate of the 1st point: '))
y1=float(input('Enter the y-coordinate of the 1st point: '))
x2=float(input('Enter the x-coordinate of the 2nd point: '))
y2=float(input('Enter the y-coordinate of the 2nd point: '))

d=((x1-x2)**2+(y1-y2)**2)**0.5

print(f'Euclidean Distance between the two coordinates: {round(d,2)}')
````

````python
def distance(x1,y1,x2,y2):
    return round(((x1-x2)**2+(y1-y2)**2)**0.5,2)
distance(4,5,2,9)
````

## Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
import math
n1= int(input("Enter numerator of first fraction: "))
d1= int(input("Enter denominator of first fraction: "))
n2= int(input("Enter numerator of second fraction: "))
d2= int(input("Enter denominator of second fraction: "))


sum_num=n1*d2+n2*d1
sum_den=d1*d2

gcd=math.gcd(sum_num, sum_den)

simplified_num = sum_num // gcd
simplified_den = sum_den // gcd

print(f"The sum of {n1}/{d1} and {n2}/{d2} is {simplified_num}/{simplified_den}")
````
