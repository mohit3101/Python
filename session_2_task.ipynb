### `Problem 11`:
A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
# direction=input('''
# Enter the direction:
# 1-> UP
# 2-> DOWN
# 3-> LEFT
# 4-> RIGHT
# 5-> STOP!

# ''')

position=[0,0]

while True:
    direction=input('''
Enter the direction:
1-> UP
2-> DOWN
3-> LEFT
4-> RIGHT
5-> STOP!

''')
    
    if direction=='1':
        step=int(input('Enter the number of steps:'))
        position[1]=position[1]+step
    elif direction=='2':
        step=int(input('Enter the number of steps:'))
        position[1]=position[1]-step
    elif direction=='3':
        step=int(input('Enter the number of steps:'))
        position[0]=position[0]-step
    elif direction=='4':
        step=int(input('Enter the number of steps:'))
        position[0]=position[0]+step
    elif direction=='5':
        break
        
distance=round((abs(position[0])**2 + abs(position[1])**2)**0.5)    
print(f'Distance travelled is: {distance}')
    
    
````

### `Problem 13`: Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
def armstrong(n):
    sum=0
    for i in list(str(n)):
        sum=sum+int(i)**3
    if n==sum:
        return True
    else:
        return False
lst=[]

start = int(input("Enter the start of the range: "))
end = int(input("Enter the end of the range: "))

for i in range(start, end+1):
    if armstrong(i):
        lst.append(i)
        
    
print(lst)    
    
````

###`Problem 14`:Calculate the angle between the hour hand and minute hand.

Input:<br>
H = 9 , M = 0<br>
Output:<br>
90<br>
Explanation:<br>
The minimum angle between hour and minute
hand when the time is 9 is 90 degress.

````python
def angle_diff(h,m):
    h_angle=h*30+0.5*m
    m_angle=m*6
    angle_diff=min(abs(h_angle-m_angle),360-abs(h_angle-m_angle))
    return round(angle_diff)

h=int(input('Enter the value of Hour hand:'))
m=int(input('Enter the value of Minute hand:'))


print(angle_diff(h,m))

````
